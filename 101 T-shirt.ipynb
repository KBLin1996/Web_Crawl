```python
import os

os.makedirs('./101 T-shirt/', exist_ok=True)

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re
import time

URL = "https://www.101.com.tw/goods-index-101.html?c_id=31100108"
html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
img_ul = soup.find_all('div', class_='rightitems')


for ul in img_ul:
    imgs = ul.find_all('img')

    for img in imgs:
        url = img["src"]
        print(url)
        
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        with open('./101 T-shirt/%s' % image_name, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1000):
                f.write(chunk)
        print('Saved %s' % image_name)
```
