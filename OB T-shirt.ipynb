```python
import os

os.makedirs('./OB T-shirt/', exist_ok=True)

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re
import time

URL = "https://www.obdesign.com.tw/inpage.aspx?no=7032&gclid=CjwKCAjw3cPYBRB7EiwAsrc-ueAyMz6TJy5PoSRoPo3Pc1zByBd7GnV9BxVELslHMgx_ivo-vTl44BoCDpkQAvD_BwE&gclsrc=aw.ds"
html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
img_ul = soup.find_all('div', id='tablenormal')


for ul in img_ul:
    imgs = ul.find_all('img')
    #print(imgs)

    for img in imgs:
        url = img["src"]
        #url = 'http:/' + url
        print(url)
        #time.sleep(60)
        
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        with open('./OB T-shirt/%s' % image_name, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1000):
                f.write(chunk)
        print('Saved %s' % image_name)
```
