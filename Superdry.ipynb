```python
import os

os.makedirs('./Superdry/', exist_ok=True)

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

for i in range(1, 25):
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(chrome_options=chrome_options)
    URL = "https://www.zalora.com.tw/men/clothing/t-shirts/special-price/?gender=men&dir=desc&sort=popularity&category_id=4219&page="
    URL = URL + str(i)
    URL = URL + "&enable_visual_sort=1"

    driver.get(URL)
    pageSource = driver.page_source
    soup = BeautifulSoup(pageSource, "lxml")
    img_ul = soup.find_all('div', class_='l-page')

    for ul in img_ul:
        imgs = ul.find_all('img')

        for img in imgs:
            url = img["src"]
            #url = "https:/" + url
            print(url)
        
            r = requests.get(url, stream=True)
            image_name = url.split('/')[-1]
            with open('./Superdry/%s' % image_name, 'wb') as f:
                for chunk in r.iter_content(chunk_size=1024):
                    f.write(chunk)
            print('Saved %s' % image_name)
```
