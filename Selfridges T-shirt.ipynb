```python
import os

os.makedirs('./Selfridges T-shirt/', exist_ok=True)

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re
import time

for i in range(1,45):
    URL = "http://www.selfridges.com/TW/en/cat/mens/clothing/tops-t-shirts/?browsing_country=TW&cm_sp=MegaMenu-_-Men-_-Clothing-Tops_tshirts&currency=TWD&language=en&pn="
    URL = URL + str(i)
    
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
    re = requests.get(URL, headers=headers) 
    re.encoding = 'utf8'

    html = requests.get(URL).text
    soup = BeautifulSoup(html, 'lxml')
    img_ul = soup.find_all('div', class_='products')
    #print(img_ul)

    for ul in img_ul:
        imgs = ul.find_all('img')
        #print(imgs)
    
        for img in imgs:
            url = img["data-mainsrc"]
            print(url)
            time.sleep(3)
            
            r = requests.get(url, stream=True)
            image_name = url.split('/')[-1]
            with open('./Selfridges T-shirt/%s' % image_name, 'wb') as f:
                for chunk in r.iter_content(chunk_size=4096):
                    f.write(chunk)
            print('Saved %s' % image_name)
```
