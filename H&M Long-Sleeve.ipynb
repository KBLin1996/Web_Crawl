```python
import os

os.makedirs('./H&M Long-Sleeve/', exist_ok=True)

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re
import shutil

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
driver = webdriver.Chrome(chrome_options=chrome_options)
URL = 'http://www2.hm.com/zh_asia3/men/shop-by-product/t-shirts-and-tanks.html?product-type=men_tshirtstanks_longsleeve%2Cmen_tshirtstanks_longsleeve&sort=stock&offset=0&page-size=90'

driver.get(URL)
pageSource = driver.page_source

soup = BeautifulSoup(pageSource, "lxml")

img_ul = soup.find_all('div', class_='productlisting parbase section')
i=0

for ul in img_ul:
    imgs = ul.find_all('img')

    for img in imgs:
        
        url = img["src"]
                
        b = 'https:'
        url = b + url
        print(url)
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        
        image_name = str(i)
        i += 1
        
        with open('./H&M Long-Sleeve/%s' % image_name, 'wb') as f:
            shutil.copyfileobj(r.raw, f)
        f.close()
        del r
```
