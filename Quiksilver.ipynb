```python
import os

os.makedirs('./Quiksilver T-Shirt/', exist_ok=True)

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re

URL = "https://www.quiksilver.com.tw/clothing/%E4%B8%8A%E8%A1%A3/t-shirts"

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
re = requests.get(URL, headers=headers) 
re.encoding = 'utf8'

html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
img_ul = soup.find_all('div', class_='category-products product-grid')


for ul in img_ul:
    imgs = ul.find_all('img')

    for img in imgs:
        url = img["src"]
        print(url)
        
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        with open('./Quiksilver T-Shirt/%s' % image_name, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1000):
                f.write(chunk)
        print('Saved %s' % image_name)

URL = "https://www.quiksilver.com.tw/clothing/%E4%B8%8A%E8%A1%A3/t-shirts?p=2"

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
re = requests.get(URL, headers=headers) 
re.encoding = 'utf8'

html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
img_ul = soup.find_all('div', class_='category-products product-grid')


for ul in img_ul:
    imgs = ul.find_all('img')

    for img in imgs:
        url = img["src"]
        print(url)
        
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        with open('./Quiksilver T-Shirt/%s' % image_name, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1000):
                f.write(chunk)
        print('Saved %s' % image_name)

URL = "https://www.quiksilver.com.tw/clothing/%E4%B8%8A%E8%A1%A3/t-shirts?p=3"

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
re = requests.get(URL, headers=headers) 
re.encoding = 'utf8'

html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
img_ul = soup.find_all('div', class_='category-products product-grid')


for ul in img_ul:
    imgs = ul.find_all('img')

    for img in imgs:
        url = img["src"]
        print(url)
        
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        with open('./Quiksilver T-Shirt/%s' % image_name, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1000):
                f.write(chunk)
        print('Saved %s' % image_name)
```
