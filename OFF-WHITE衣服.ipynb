```python
import os

os.makedirs('./Off-White T-shirt/', exist_ok=True)
```

```python
# 爬OFF-WHITE的T-shirt
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re

URL = "http://www.selfridges.com/TW/en/cat/mens/clothing/tops-t-shirts/?cm_sp=MegaMenu-_-Men-_-Clothing-Tops_tshirts"
html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
img_ul = soup.find_all('div', class_='productsInner')

for ul in img_ul:
    imgs = ul.find_all('img')
    
    for img in imgs:
        url = img["data-mainsrc"]
        print(url)
        
        r = requests.get(url, stream=True)
        image_name = url.split('/')[-1]
        with open('./Off-White T-shirt/%s' % image_name, 'wb') as f:
            for chunk in r.iter_content(chunk_size=128):
                f.write(chunk)
        print('Saved %s' % image_name)
```
